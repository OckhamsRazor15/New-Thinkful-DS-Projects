```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
raw_data = pd.read_csv('epi_r.csv')
```

```python
from sklearn.svm import SVR
svr = SVR()
X = raw_data.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1).dropna(axis=1)
Y = raw_data.rating
#svr.fit(X,Y)
```

```python
svr.score(X, Y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svr, X, Y, cv=5)
```

```python
#MY CODE
#let's create a function to make this easy for us
def myFunc(x):
    if x == 5:
        return 5
    elif x == 0:
        return 0
    elif x >= 4:
        return 4
    elif x >= 3:
        return 3
    elif x >= 2:
        return 2
    else:
        return 1
    
raw_data["rating2"] = raw_data["rating"].apply(myFunc)


#second one with only 2 categories
def myFunc2(x):
    if x >= 2.5:
        return 1
    else:
        return 0

raw_data["rating3"] = raw_data["rating"].apply(myFunc2)
```

```python
#per the reqs, we need to widdle it down to 30 most important. we only care about predictive power, so PCA is a good fit here.
from sklearn.decomposition import PCA
pca = PCA(n_components=30)

X_pca = pca.fit_transform(X)
df2 = pd.DataFrame({"1": X_pca[:,0],"2": X_pca[:,1],"3": X_pca[:,2],"4": X_pca[:,3],
                   "5": X_pca[:,4],"6": X_pca[:,5],"7": X_pca[:,6],"8": X_pca[:,7],
                   "9": X_pca[:,8],"10": X_pca[:,9],"11": X_pca[:,10],"12": X_pca[:,11],
                   "13": X_pca[:,12],"14": X_pca[:,13],"15": X_pca[:,14],
                   "16": X_pca[:,15],"17": X_pca[:,16],"18": X_pca[:,17],"19": X_pca[:,18],
                   "20": X_pca[:,19],"21": X_pca[:,20],"22": X_pca[:,21],"23": X_pca[:,22],
                   "24": X_pca[:,23],"25": X_pca[:,24],"26": X_pca[:,25],"27": X_pca[:,26],
                   "28": X_pca[:,27],"29": X_pca[:,28],"30": X_pca[:,29]})
```

```python
from sklearn.svm import SVR
svr = SVR()
X = df2
Y = raw_data.rating3
svr.fit(X,Y)
```

```python
#see the score + make sure there is no overfitting
from sklearn.model_selection import cross_val_score
cross_val_score(svr, X, Y, cv=5)
```
