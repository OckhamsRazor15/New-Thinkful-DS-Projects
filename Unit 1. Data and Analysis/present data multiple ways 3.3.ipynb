```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#%matplotlib inline
```

```python
crashes = pd.read_csv("plane_crashes_data.csv")
```

```python
crashes.head()
```

## 1. 1 variable

```python
sns.distplot(crashes["fatalities"], bins=10)
```

```python
sns.rugplot(crashes["fatalities"])
```

```python
plt.plot(crashes["fatalities"])
```

## 2. 2 continous variables

```python
sns.jointplot(x="aboard",y="fatalities",data=crashes, kind="hex")
```

```python
sns.boxplot(x='aboard',y='fatalities',data=crashes)
```

```python
plt.scatter(crashes["aboard"], crashes["fatalities"])
```

## 3. 1 continous and 1 categorical

```python
plt.scatter(crashes["month"], crashes["fatalities"])
```

```python
sns.barplot(x=crashes["month"], y=crashes["fatalities"], data=crashes)
```

```python
plt.plot(crashes["month"], crashes["fatalities"])
```

```python
sns.boxplot(crashes["month"], crashes["fatalities"])
```

```python
sns.violinplot(crashes["month"], crashes["fatalities"])
```

```python
sns.pointplot(crashes["month"], crashes["fatalities"])
```

### While it's easy to use any plot, it's obvious from above that the plot that you choose must best represent your data. The line chart, especially in the 1 continous and 1 categorical, is a mess. The violin plot or the box plot tell a much better story. The joint plot looks cool, but with our data being so varied, the hexes get lost in the negative space.
